```python
import pandas as pd
data = pd.read_excel('wheather.xls')  # Считываем Excel-таблицу в переменную data
data.head(10)  # Чтение 10 первых строк
```

```python
data = pd.read_excel('wheather.xls', skiprows = 6)  
data
```

```python
data.columns
```

```python
data['T']  # Рассмотрение столбцов температуры
```

```python
data['Местное время в Москве (ВДНХ)'][5]  # Нужно преобразовать эти данные, т.к. Python не умеет с ними работать
```

```python
pd.to_datetime(data['Местное время в Москве (ВДНХ)'], dayfirst=True)  # dayfirst=True, т.к. формат дней Российский, а не Американский
```

```python
data['date'] = pd.to_datetime(data['Местное время в Москве (ВДНХ)'], dayfirst=True)  # Для удобства создадим новую колонку date
plt.figure(figsize=(20, 5))
plt.plot(data['date'], data['T'], color = 'red')  # В качестве x - колонка с датой, y - колонка с температурой
```

```python
condition = data['T'].isna()  # Проверка на пустые данные
data[condition] 
```

```python
data['T'].hist () # Строим гистограмму для температуры. Слева - сколько раз встречалась, снизу - значение температуры
```

```python
data['T'].quantile(0.85)  # В 85% случаев температура составляет...(-0.7)
```

```python
condition = data['date'] < '05-02-2022'
data_short = data[condition]

plt.figure(figsize=(20,5))
plt.plot(data_short['date'], data_short['T'])
```
